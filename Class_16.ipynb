```python
df=df[df.claim_amt>350]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.reindex(index=['a','b','c','d','e','f'],columns=[''])
```

```python
df.reindex()
```

```python
df.reindex(index=[1,2,3,4],columns=['Insured_id','proc_cd'])
```

```python
df.rename(columns={'Insured_id':'Ins_id'},index={'a':'aa'},inplace=True)
```

```python
pd.DataFrame(df.groupby('Ins_id').agg('mean').reset_index())
```
